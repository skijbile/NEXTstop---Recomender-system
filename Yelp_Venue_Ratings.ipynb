```python
import pandas as pd
import json
```

```python
with open('ratings_yelp1.json','r') as f:
    ratings_yelp1=json.load(f)
with open('ratings_yelp2.json','r') as f:
    ratings_yelp2=json.load(f)
```

```python
Business_df=pd.read_csv('Business_info.csv')
Business_df
```

```python
"""
columns required from the json
reviews-text, rating
"""
```

```python
#ratings_yelp1[1]see the format of json
```

```python
review=[]
ratings=[]
venue_name=[]
venue_id=[]
for index, row in Business_df.iterrows():
    json_item1=ratings_yelp1[index]
    json_item2=ratings_yelp2[index]
    
    if 'reviews' in json_item1:
        for review_item in json_item1['reviews']:

            if 'text' in review_item:
                review.append(review_item['text'])
            else:
                review.append('Not available')

            if 'rating' in review_item:
                ratings.append(review_item['rating'])
            else:
                ratings.append('Not available')
                
            venue_name.append(row.iloc[1])
            venue_id.append(row.iloc[0])
    else:
        review.append('Not available')
        ratings.append('Not available')
        venue_name.append(row.iloc[1])
        venue_id.append(row.iloc[0])

    # for 'reviews' in json_item2:
    #     reviews.append(item['text'])
    #     ratings.append(item['rating'])
    #     venue_name.append(row.iloc[1])
    #     venue_id.append(row.iloc[0])

df=pd.DataFrame({'venue':venue_name,
                 'venue_id':venue_id,
                 'reviews':review,
                 'ratings':ratings})
df
    
```

```python
df.to_csv('Venue_Ratings.csv',index=False)
```
