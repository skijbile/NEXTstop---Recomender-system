## Event Search for Canada
#### use the URL to get events based on the query parameter of countryCode-Canada

```python
# define variables for the required data
country_code="CA" # code for Canada to get events
endpoint= '/discovery/v2/events' #endpoint to get event_ID
segment =["Arts"] # use different segments (music,arts,sports,theatre)
```

```python
URL = f"https://app.ticketmaster.com{endpoint}.json?countryCode={country_code}&apikey={TicketMaster_key}&segmentName={segment}"
print(URL)
```

```python
res= get_events(endpoint,TicketMaster_key,country_code,segment,0) # initial page number is 0
res.status_code
```

```python
events=[] # empty list to add each json file as an object
counter = 0  # to keep track of iterations

# to give correct outputs and save the data as json file 
# call the function to get data

if res.status_code==200:
        events_data= res.json()
        if 'page' in events_data and 'totalPages' in events_data['page']:
            total_pages = events_data['page']['totalPages']
    
            for page_number in range(0,total_pages):
                response = get_events(endpoint,TicketMaster_key, country_code, segment, page_number)
                data = response.json()

                if 'events' in data["_embedded"]:
                    events.extend(data['_embedded']['events']) # adding data to existing list (empty or old data)
                 # Append events to the JSON file
                    #change file name for different parameter setting for segment or country etc.
                    with open(f'Events_all_2.json', 'a') as f: 
                        json.dump(events, f) 
                        f.write('\n')  # Add a newline character after each JSON object
                    counter+=1 
                    if counter%100==0: # to track progress if required
                        print("Rows processed = ", counter)
                else:
                    print(f"Error: No events found on page {page_number}")

            #  # Write all events to the JSON file
            # # change json to an append mode 'a' to to add events from each segment without overwriting the existing data.
            # with open('Events_all_Canada.json', 'a') as f:
            #     json.dump(events, f)

        print("END of list") # to know end of iteration
        
else:
    print(f"Error: {res.status_code}, {res.text}")
    

  
    
```
